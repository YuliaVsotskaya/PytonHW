```python
Задание 1
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
Под временем жизни понимается разница между максимальным 
и минимальным значением столбца timestamp для данного значения userId
```

```python
df = pd.read_csv('ratings.dat', names = ['UserIDs', 'MovieIDs', 'Rating', 'Timestamp'], sep='::', header=None)
df.head()
```

```python
df_count = df.groupby('UserIDs').count()[['Rating']]
df_users = df_count.loc[df_count.Rating > 100]
```

```python
df_time_max = df.groupby('UserIDs').max()[['Timestamp']].rename(columns={'Timestamp':'time_max'})
df_time_min = df.groupby('UserIDs').min()[['Timestamp']].rename(columns={'Timestamp':'time_min'})
```

```python
df_times = df_users.join(df_time_max).join(df_time_min)
df_times['time_avg'] = df_times['time_max'] - df_times['time_min']
```
